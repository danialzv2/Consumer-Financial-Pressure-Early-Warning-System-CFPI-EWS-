```python
import pandas as pd
import numpy as np
import requests
import pprint
```

```python
def json_to_df(data):
    if isinstance(data, list):
        if all(isinstance(item, dict) for item in data):
            return pd.json_normalize(data)
        else:
            return pd.DataFrame({"value": data})

    elif isinstance(data, dict):
        return pd.json_normalize(data)

    else:
        return pd.DataFrame({"value": [data]})
```

# Fuel Data Collection using DOSM API

```python
url = "https://api.data.gov.my/data-catalogue?id=fuelprice" 
response_json = requests.get(url=url).json()
fuel_data = json_to_df(response_json)
fuel_data.head()

fuel_data.to_csv("../raw_data/fuel_data.csv")
```

# Inflation Data Collection using DOSM API

```python
url = "https://api.data.gov.my/data-catalogue?id=cpi_state&divison=overall"
response_json = requests.get(url=url).json()
inflation_data = json_to_df(response_json)
inflation_data.head()

inflation_data.to_csv("../raw_data/inflation_data.csv")
```

# OPR Data Collection using BNM OpenAPI

### Real Time Data Collection

```python
import requests

url = "https://api.bnm.gov.my/public/opr"

headers = {
    "Accept": "application/vnd.BNM.API.v1+json"
}

response = requests.get(url, headers=headers)

# Check status
print("Status code:", response.status_code)

# JSON output
data = response.json()
loan_approval = json_to_df(data)
loan_approval.head()
```
